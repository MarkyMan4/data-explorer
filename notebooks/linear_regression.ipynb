```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data_files/iris_data.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()

df = df.sort_values(df.columns[2])

mask = np.random.randn(len(df)) < 0.8

train = df[mask]
test = df[~mask]

train = train.sort_values(df.columns[2])
test = test.sort_values(df.columns[2])

x = np.array(train[[df.columns[2]]])
y = np.array(train[df.columns[3]])

test_x = np.array(test[[df.columns[2]]])
test_y = np.array(test[df.columns[3]])

lin_reg.fit(x, y)
```

```python
pred = lin_reg.predict(test_x)
```

```python
fig = plt.figure(figsize=(15,8))
raw = plt.scatter(x, y, color='red')
fit, = plt.plot(test_x, pred, linewidth=3, color='blue')
plt.legend((raw, fit), ('Data', 'Linear Fit'), loc='lower right')
plt.title('Linear regression')
plt.show()
```

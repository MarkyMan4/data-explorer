# Template for linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import operator
```

```python
df = pd.read_csv('data_files/iris_data.csv')
df
```

```python
df = df.sort_values(df.columns[1])

x = np.array(df[[df.columns[1]]])
y = np.array(df[[df.columns[0]]])

# x = np.sort(x)

# add an additional axis (necessary for polynomial regression)
# x = x[:, np.newaxis]
# y = y[:, np.newaxis]
```

```python
poly_features = PolynomialFeatures(degree=15)
x_poly = poly_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)
```

```python
fig = plt.figure(figsize=(15,8))
raw = plt.scatter(x, y, color='red')
fit, = plt.plot(x, y_poly_pred, linewidth=3, color='blue')
plt.legend((raw, fit), ('Data', 'Polynomial Fit'), loc='lower right')
plt.title('Polynomial regression')
plt.show()
```
